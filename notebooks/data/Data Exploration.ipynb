```python
y_ds.describe()
```

```python
wv = x_ds.columns[12]
xwv = x_ds.loc[:, wv].copy()
xwv = xwv / xwv.abs().sum()* 100
xwv.plot.hist(bins=20)
mu = xwv.mean()
sigma = xwv.std()
b = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(b, stats.norm.pdf(b, mu, sigma))

plt.show()
```

```python
p = y_ds.columns[2]
xwv = y_ds[p].copy()
xwv = xwv / xwv.abs().sum()* 100
bins = int(30*(y_ds[p].max() - y_ds[p].min()) /y_ds[p].std())
mu = np.log(pigment_hist.loc[pigment_trad[p]]['Average (ng/l)']/1000) if p in pigment_trad.keys() else 0
sigma = pigment_hist.loc[pigment_trad[p]]['SD (ng/l)']/1000 if p in pigment_trad.keys() else 0
b = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(b, stats.norm.pdf(b, mu, sigma))
xwv.plot.hist(bins=bins)
plt.plot(True)
plt.show()
```
